```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Tugas Mosi 2 Week 3
# Gerak Jatuh Bebas

N = 21
T = 5
v = 0
G = 9.8
y = [0 for i in range(N)]
dt = T/(N)
t = np.arange(0,T,dt)

#Solusi Numerik
y[0] = 100
# Rumus y[1] jika i=1,2,..,N
y[1] = v*dt+y[0]

for i in range(2,N):
    # Dari rumus
    y[i] = (-G*(dt**2)+(2*y[i-1])-y[i-2])

#Solusi eksak
'''
N=21
y(0)=100

v=0/s
G=9.8m/s^2
T=0
dt=0.05 (Dari rumus 1/(N-1))

saat t=0
v=-G*t+c1
1=-9.8*0+c1
c1=1

saat t=0
y(0)=-1/2*G*t^2+c1*t+c2
100=0+0+c2
c2=100

'''
c1=1
c2=100

f = []
for i in t:
    f.append((-(G*(i**2))/2)+c1*i+c2)
    
    
# Perbandingan numerik(Merah) dan eksak(Biru)
sns.pointplot(t[:N],f[:N],figsize=(20,7))
sns.pointplot(t[:N],y,figsize=(20,7),color="R")
```
